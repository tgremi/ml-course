```python
db.columns
```

```python
colunasUsadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
colunasUsadas
```

```python
base = pd.read_csv('original.csv', usecols = colunasUsadas)
```

```python
base.head()
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler_x = MinMaxScaler()
```

```python
itensNormalize = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 
                  'grade','sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long']
```

```python
base[itensNormalize] = scaler_x.fit_transform(base[itensNormalize])
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
x = base.drop('price', axis = 1)
y = base.price
```

```python
x.head()
```

```python
y.head()
```

```python
previsoresColunas = itensNormalize
previsoresColunas
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
colunas = [tf.feature_column.numeric_column(key = c) for c in previsoresColunas]
```

```python
from sklearn.model_selection import train_test_split
xTreinamento, xTeste, yTreinamento, yTeste = train_test_split(x, y, test_size = 0.3)
```

```python
funcaoTreinamento = tf.estimator.inputs.pandas_input_fn(x = xTreinamento, y = yTreinamento, batch_size = 32, 
                                                        num_epochs = None, shuffle = True)
```

```python
funcaoTeste =  tf.estimator.inputs.pandas_input_fn(x = xTeste, y = yTeste, batch_size = 32, 
                                                               num_epochs = 10000, shuffle = False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
regressor.train(input_fn = funcaoTreinamento, steps = 10000)
```

```python
metricasTreinamento = regressor.evaluate(input_fn = funcaoTreinamento, steps = 10000)
```

```python
metricasTeste = regressor.evaluate(input_fn = funcaoTeste, steps = 10000)
```

```python
metricasTeste
```

```python
metricasTreinamento
```

```python
funcaoPrevisao = tf.estimator.inputs.pandas_input_fn(x = xTeste, shuffle=False)
```

```python
previsoes = regressor.predict(input_fn=funcaoPrevisao)
```

```python
list(previsoes)
```

```python
valoresPrevisoes = []

for p in regressor.predict(input_fn=funcaoPrevisao):
    valoresPrevisoes.append(p['predictions'])
```

```python
valoresPrevisoes
```

```python
import numpy as np 
valoresPrevisoes = np.asarray(valoresPrevisoes).reshape(-1,1)
```

```python
valoresPrevisoes
```

```python
valoresPrevisoes = scaler_y.inverse_transform(valoresPrevisoes)
```

```python
valoresPrevisoes
```

```python
yTeste2 = yTeste.values.reshape(-1, 1)
```

```python
yTeste2 = scaler_y.inverse_transform(yTeste2)
yTeste2
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(yTeste2, valoresPrevisoes)
```
